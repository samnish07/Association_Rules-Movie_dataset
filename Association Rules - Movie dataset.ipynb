## Association rules with 5% support and 20% confidence

```python
# with 5% support
frq_items = apriori(movies_enc, min_support = 0.05, use_colnames = True)
frq_items['length'] = frq_items['itemsets'].apply(lambda x: len(x))
frq_items
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

frq_items_sort = frq_items.sort_values(by='support',ascending=False)
frq_items_sort['itemsets'] = frq_items_sort['itemsets'].apply(lambda a: ','.join(list(a)))
plt.figure(figsize = (16,16))
ax = sns.barplot(x = 'support', y = 'itemsets', data=frq_items_sort)
plt.yticks(rotation = 0, fontsize = 14)
plt.xticks(rotation = 45, fontsize = 12)
plt.title('Most frequently watched movies',  fontsize = 18, fontweight = 'bold')
for i in ax.containers:
    ax.bar_label(i,)
```

```python
frq_items_filter1 = frq_items[frq_items.support > 0.5]
frq_items_sort = frq_items_filter1.sort_values(by='support',ascending=False)
frq_items_sort1 = frq_items_sort.copy()
frq_items_sort1['itemsets'] = frq_items_sort['itemsets'].apply(lambda a: ','.join(list(a)))
plt.figure(figsize = (7,3))
ax = sns.barplot(x = 'support', y = 'itemsets', data=frq_items_sort1)
plt.yticks(rotation = 0, fontsize = 14)
plt.xticks(rotation = 0, fontsize = 12)
plt.title('Most frequently watched movies',  fontsize = 18, fontweight = 'bold')
for i in ax.containers:
    ax.bar_label(i,)
```

###### ***The movies Gladiator, Sixth Sense and Patriot are the most frequently watched ones. Also the support value for the set Gladiator and Patriot is significant.***

```python
# with 20% confidence
rules_20 = association_rules(frq_items,metric='confidence', min_threshold=0.2)
rules_20_sort = rules_20.sort_values(by='confidence',ascending=False)
rules_20_sort
```

```python
pd.set_option('display.max_rows',None)
rules_20_filter = rules_20[rules_20.confidence>0.8]
rules_20_filter
```

```python
rules_20_sorted = association_rules(frq_items_sort,metric='confidence', min_threshold=0.2)
rules_20_sorted
```

#### ***Gladiator and Patriot set has 60% support, 85.7% and 100% confidence and lift value slightly > 1***

```python
rules_20_1 = rules_20.copy()
rules_20_1["antecedents"] = rules_20["antecedents"].apply(lambda x: ','.join(list(x)))
rules_20_1["consequents"] = rules_20["consequents"].apply(lambda x: ','.join(list(x)))
rules_20_lift = rules_20_1[rules_20_1.lift > 1.0]
rules_20_lift_sort = rules_20_lift.sort_values(by='lift',ascending=False)
rules_20_lift_sort
```

#### Confidence says how likely item y is purchased when item x is purchased, expressed as {x->y} . However, one drawback of confidence measure is it does not account for the base popularity of both constituent items individually. So a third measure called Lift is used.

#### Lift says how likely item y is purchased when item x is purchased while accounting for how popular y is.
#### - Lift = 1 --> no association between items
#### - Lift > 1 --> item y is likely to be bought if item x is bought
#### - Lift < 1 --> item y is unlikely to be bought if item x is bought


```python
sns.scatterplot(x='support',y='confidence',data=rules_20);
```

```python
matrix = rules_20_lift_sort.pivot('antecedents','consequents','lift')
matrix
```

```python
plt.figure(figsize=(20,6),dpi=250)
sns.heatmap(matrix,annot=True)
plt.title('HeatMap - ForLiftMatrix')
plt.yticks(rotation=0)
plt.xticks(rotation=90);
```
